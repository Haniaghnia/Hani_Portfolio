4.2.1 Visualization Categorical Feature

```python
auto_categorical = auto.describe(include=['object'])
auto_categorical.T
```

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

# Iterate over categorical features and plot their distributions
row = 0
col = 0
for feature in auto_categorical:
    value_counts = auto[feature].value_counts()

    # Check if row and col are within the bounds of the axes array
    if row >= axes.shape[0] or col >= axes.shape[1]:
        break

    # Get the appropriate axes for plotting
    ax = axes[row, col]

    # Create a bar plot
    value_counts.plot(kind='bar', ax=ax)

    # Set plot title, labels, and adjust layout
    ax.set_title(f"Frequency Distribution of {feature}")
    ax.set_xlabel(feature)
    ax.set_ylabel("Count")

    # Move to the next column or row
    col += 1
    if col == 3:
        col = 0
        row += 1

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

### 6. Descriptive Statistical Analysis

#### 6.1 Correlation and Causation

P-value:

The P-value is the probability value that the correlation between these two variables is statistically significant. Normally, we choose a significance level of 0.05, which means that we are 95% confident that the correlation between the variables is significant.

By convention, when the

- p-value is  < 0.001: we say there is strong evidence that the correlation is significant.
- the p-value is  < 0.05: there is moderate evidence that the correlation is significant.
- the p-value is  < 0.1: there is weak evidence that the correlation is significant.
- the p-value is  > 0.1: there is no evidence that the correlation is significant.

We can obtain this information using "stats" module in the "scipy" library.

```python
auto_num_corr = auto_num.corr()['price']
golden_features_list = auto_num_corr[abs(auto_num_corr)> 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with Price:\n{}".format(len(golden_features_list), golden_features_list))
```

we can use the function 'f_oneway' in the module `'stats' `to obtain the `F-test `score and P-value.

```python
# ANOVA
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'],
                             grouped_test2.get_group('rwd')['price'],
                             grouped_test2.get_group('4wd')['price'])
print('ANOVA results : F=', f_val, ',P=', p_val)
```

The f-statistic is higher indicating a greater difference between group means. The result is a statistically significant difference between the means of group Drive Wheels.

**Fwd and Rwd**

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'],
                              grouped_test2.get_group('rwd')['price'])
print('ANOVA result : F= ', f_val, ',P=', p_val)
```

The f-statistic is higher indicating a greater difference between group means. The result is a statistically significant difference between the means `fwd` and `rwd`

**4wd and rwd**

```python
f_val , p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'],
                               grouped_test2.get_group('rwd')['price'])
print('ANOVA result : F= ', f_val, ',P=',p_val)
```

This means that there is a statistically significant difference between the means of the groups. The p-value is less than 0.05, which is the commonly used threshold for statistical significance.
But the F-statistic is 8.58, which is considered a moderate difference between `4wd` and rwd


**4wd and fwd**

```python
f_val , p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'],
                               grouped_test2.get_group('fwd')['price'])
print('ANOVA result : F= ', f_val, ',P=',p_val)
```

This means that there is not a statistically significant difference between the means of the groups. The p-value is greater than 0.05, which is the commonly used threshold for statistical significance.

A higher F-statistic indicates a greater difference between group means. But in this case, the F-statistic is 0.665, which is very low and indicates a very low difference between group mean `4wd` and `fwd`
